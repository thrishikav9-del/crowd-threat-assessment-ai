```python
# ══════════════════════════════════════════════════════════════════
# CELL 2 – All imports, global constants, and model definitions
# FIXED: ThreatNet uses LayerNorm (not BatchNorm1d) to avoid
#        ValueError when batch size == 1 at end of epoch.
# ══════════════════════════════════════════════════════════════════
import cv2, os, glob, json, pickle, shutil, warnings, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import h5py
warnings.filterwarnings("ignore")

from pathlib import Path
from tqdm.auto import tqdm
from ultralytics import YOLO
from sklearn.neighbors import NearestNeighbors
import supervision as sv
import torchvision.transforms as T
from torchvision import transforms
from torchvision.models import resnet18, ResNet18_Weights
from transformers import ViTMAEForPreTraining
from google.colab.patches import cv2_imshow
from PIL import Image

# ── Device & global paths ──────────────────────────────────────────
DEVICE         = "cuda" if torch.cuda.is_available() else "cpu"
IMG_SIZE       = (320, 240)
CONF_THRESHOLD = 0.18

FLAT_DIR      = "/content/shanghaitech_flat_images"
HDF5_OUTPUT   = "/content/shanghaitech_phase1.h5"
SUMMARY_CSV   = "/content/image_summary.csv"
ENHANCED_CSV  = "/content/image_summary_enhanced.csv"
MANIFEST_PATH = "/content/top50_anomaly_manifest.json"
ANOMALY_PATH  = "/content/anomaly_trigger_mae.pth"
THREAT_PATH   = "/content/threat_value_net.pth"
POLICY_PATH   = "/content/policy_priors_manual.pkl"
BN_PATH       = "/content/bayesian_network.pkl"

# ── Model architecture (✅ Improvement #3: clear description) ──────

class ThreatNet(nn.Module):
    """
    Lightweight MLP threat scorer.

    Architecture:
        Input (6-dim)
          → Linear(6→64) → LayerNorm(64) → ReLU → Dropout(0.2)
          → Linear(64→32) → LayerNorm(32) → ReLU
          → Linear(32→1) → Sigmoid × 10  →  ̂t ∈ [0, 10]

    NOTE: LayerNorm replaces BatchNorm1d — safe for batch_size=1.
    """
    def __init__(self, input_dim: int = 6, hidden_dim: int = 64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.LayerNorm(hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, 1),
            nn.Sigmoid()
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.net(x) * 10.0


class AnomalyTrigger(nn.Module):
    """
    ViT-MAE encoder + binary classification head.

    Architecture:
        Pixel values (3×224×224)
          → ViT-MAE encoder (12 layers, hidden=768)
          → CLS token z_CLS ∈ R^768
          → Linear(768→256) → LayerNorm(256) → GELU → Dropout(0.3)
          → Linear(256→1) → Sigmoid  →  â ∈ [0,1]

    Gradual fine-tuning: encoder frozen for epochs 1–6, unfrozen 7–12.
    Gate threshold τ_a = 0.7.
    """
    def __init__(self, vit_encoder, hidden_dim: int = 256):
        super().__init__()
        self.vit_encoder = vit_encoder
        encoder_dim      = vit_encoder.config.hidden_size  # 768

        self.classifier = nn.Sequential(
            nn.Linear(encoder_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.GELU(),
            nn.Dropout(0.3),
            nn.Linear(hidden_dim, 1),
            nn.Sigmoid()
        )

    def forward(self, pixel_values: torch.Tensor) -> torch.Tensor:
        out = self.vit_encoder(pixel_values=pixel_values)
        cls = out.last_hidden_state[:, 0]  # CLS token: (B, 768)
        return self.classifier(cls)        # (B, 1)


# ── Base models ────────────────────────────────────────────────────
yolo = YOLO("yolo11n.pt")

resnet    = resnet18(weights=ResNet18_Weights.DEFAULT).to(DEVICE)
resnet.fc = nn.Identity()
resnet.eval()

preprocess = T.Compose([
    T.ToPILImage(),
    T.Resize(256),
    T.CenterCrop(224),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

print(f"Device : {DEVICE}")
print("YOLO   : yolo11n.pt loaded")
print("ResNet : ResNet-18 (512-dim) loaded")
print("Cell 2 complete.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 3 – Utility functions
# ══════════════════════════════════════════════════════════════════

def build_interaction_graph(
    positions: np.ndarray,
    r: float = 60.0,
    sigma: float = 35.0
) -> np.ndarray:
    """Gaussian-weighted adjacency matrix between person centres."""
    n = len(positions)
    if n <= 1:
        return np.zeros((n, n), dtype=np.float32)
    dists = np.linalg.norm(
        positions[:, None, :] - positions[None, :, :], axis=-1
    )
    adj = np.exp(-dists ** 2 / (2 * sigma ** 2))
    adj[dists > r] = 0.0
    np.fill_diagonal(adj, 0.0)
    return adj.astype(np.float32)


def extract_resnet_features(
    frame_rgb: np.ndarray,
    bboxes: np.ndarray
) -> np.ndarray:
    """512-dim ResNet-18 features for each detected person crop."""
    if len(bboxes) == 0:
        return np.zeros((0, 512), dtype=np.float32)

    crops = []
    h, w  = frame_rgb.shape[:2]
    for box in bboxes:
        x1 = max(0, int(box[0])); y1 = max(0, int(box[1]))
        x2 = min(w, int(box[2])); y2 = min(h, int(box[3]))
        if (x2 - x1) < 4 or (y2 - y1) < 4:
            continue
        crops.append(preprocess(frame_rgb[y1:y2, x1:x2]))

    if not crops:
        return np.zeros((0, 512), dtype=np.float32)

    batch = torch.stack(crops).to(DEVICE)
    with torch.no_grad():
        feats = resnet(batch).cpu().numpy()
    return feats.astype(np.float32)


def make_feature_vector(
    centers:      np.ndarray,
    confs:        np.ndarray,
    graph:        np.ndarray,
    resnet_feats: np.ndarray,
    img_size:     tuple = IMG_SIZE
) -> np.ndarray:
    """
    6-dim feature vector for ThreatNet:
        [mean_resnet_activation, mean_graph_edge_weight, density_bin,
         mean_cx, mean_cy, mean_confidence]
    """
    n           = len(centers)
    feat_mean   = float(resnet_feats.mean()) if resnet_feats.size > 0 else 0.0
    graph_embed = float(graph.mean())        if graph.size > 0        else 0.0
    density_bin = n / (img_size[0] * img_size[1] * 0.001)
    cx          = float(centers[:, 0].mean()) if n > 0    else 0.0
    cy          = float(centers[:, 1].mean()) if n > 0    else 0.0
    conf_mean   = float(confs.mean())          if len(confs) > 0 else 0.0
    return np.array(
        [feat_mean, graph_embed, density_bin, cx, cy, conf_mean],
        dtype=np.float32
    )

print("Cell 3 complete — utility functions defined.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 4 – Process all flattened images → HDF5
# ══════════════════════════════════════════════════════════════════
IMAGE_DIR = Path(FLAT_DIR)
images    = sorted(
    list(IMAGE_DIR.glob("*.jpg")) + list(IMAGE_DIR.glob("*.jpeg"))
)

if not images:
    raise FileNotFoundError("No images found — run Cell 1 first.")

print(f"Processing {len(images)} images → {HDF5_OUTPUT}")
skipped = 0

def process_one_image(img_path: Path, h5: h5py.File) -> bool:
    key = img_path.stem
    img = cv2.imread(str(img_path))
    if img is None:
        return False

    img_rgb     = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img_rgb, IMG_SIZE)

    results = yolo(img_resized, verbose=False, imgsz=640, conf=CONF_THRESHOLD)[0]
    dets    = sv.Detections.from_ultralytics(results)
    dets    = dets[dets.class_id == 0]

    n = len(dets)
    if n == 0:
        return False

    centers      = (dets.xyxy[:, :2] + dets.xyxy[:, 2:]) / 2.0
    bboxes       = dets.xyxy
    graph        = build_interaction_graph(centers)
    resnet_feats = extract_resnet_features(img_resized, bboxes)

    g = h5.require_group(key)
    for ds_name, data in [
        ("centers",           centers),
        ("bboxes",            bboxes),
        ("confidences",       dets.confidence),
        ("interaction_graph", graph),
        ("resnet_features",   resnet_feats),
    ]:
        g.create_dataset(ds_name, data=data, compression="gzip", compression_opts=4)
    return True

with h5py.File(HDF5_OUTPUT, "w") as h5:
    pbar = tqdm(images, desc="Building HDF5")
    for p in pbar:
        found = process_one_image(p, h5)
        if not found:
            skipped += 1
        pbar.set_postfix(skipped=skipped)

print(f"Done. Skipped (0 detections): {skipped}")
print("Cell 4 complete.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 6 – Compute per-image anomaly indicators
# ══════════════════════════════════════════════════════════════════

def compute_image_metrics(h5_grp) -> dict:
    empty = dict(
        num_persons=0, density=0.0, avg_degree=0.0,
        max_degree=0, connectivity_ratio=0.0, irregularity_score=0.0
    )
    if "centers" not in h5_grp:
        return empty

    centers = h5_grp["centers"][:]
    n       = len(centers)
    if n == 0:
        return empty

    graph           = h5_grp["interaction_graph"][:]
    degrees         = graph.sum(axis=1)
    density         = n / (IMG_SIZE[0] * IMG_SIZE[1]) * 10_000
    avg_degree      = float(degrees.mean())
    max_degree      = int(degrees.max())
    connected_nodes = float((degrees > 0).sum() / n)
    degree_std      = float(degrees.std())
    irregularity    = (degree_std / (avg_degree + 1e-6)) * (1.0 - connected_nodes + 1e-6)

    return dict(
        num_persons        = n,
        density            = round(density, 4),
        avg_degree         = round(avg_degree, 4),
        max_degree         = max_degree,
        connectivity_ratio = round(connected_nodes, 4),
        irregularity_score = round(irregularity, 6),
    )

records = []
with h5py.File(HDF5_OUTPUT, "r") as h5:
    for key in tqdm(h5.keys(), desc="Computing metrics"):
        row = compute_image_metrics(h5[key])
        row["image_key"] = key
        records.append(row)

df = pd.DataFrame(records)
df.to_csv(SUMMARY_CSV, index=False)
print(f"Saved {len(df)} records → {SUMMARY_CSV}")
print(df.describe().round(3))
print("Cell 6 complete.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 7 – Visualise detections + interaction graph
# ══════════════════════════════════════════════════════════════════

def visualize_image(key: str) -> None:
    img_path = Path(FLAT_DIR) / f"{key}.jpg"
    if not img_path.exists():
        img_path = Path(FLAT_DIR) / f"{key}.jpeg"
    if not img_path.exists():
        print(f"  Image not found: {key}")
        return

    img         = cv2.cvtColor(cv2.imread(str(img_path)), cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img, IMG_SIZE)

    with h5py.File(HDF5_OUTPUT, "r") as h5:
        if key not in h5 or "bboxes" not in h5[key]:
            print(f"  No HDF5 data: {key}")
            return
        g       = h5[key]
        bboxes  = g["bboxes"][:]
        confs   = g["confidences"][:]
        centers = g["centers"][:]
        graph   = g["interaction_graph"][:]

    overlay = img_resized.copy()
    n       = len(bboxes)

    # Draw interaction edges
    for i in range(n):
        for j in range(i + 1, n):
            if graph[i, j] > 0.1:
                alpha = float(graph[i, j])
                pt1   = tuple(map(int, centers[i]))
                pt2   = tuple(map(int, centers[j]))
                col   = (int(255 * alpha), int(200 * alpha), 0)
                cv2.line(overlay, pt1, pt2, col, 1, cv2.LINE_AA)

    # Draw bounding boxes + confidence
    for box, conf in zip(bboxes, confs):
        x1, y1, x2, y2 = map(int, box)
        cv2.rectangle(overlay, (x1, y1), (x2, y2), (220, 60, 60), 1)
        cv2.putText(overlay, f"{conf:.2f}", (x1, max(y1 - 3, 8)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.32, (255, 220, 0), 1, cv2.LINE_AA)

    # Adjacency matrix
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    axes[0].imshow(overlay)
    axes[0].set_title(f"{key} — {n} persons detected", fontsize=11)
    axes[0].axis("off")
    im = axes[1].imshow(graph, cmap="plasma", vmin=0, vmax=1)
    axes[1].set_title("Interaction Graph (adjacency)", fontsize=11)
    axes[1].set_xlabel("Person index"); axes[1].set_ylabel("Person index")
    plt.colorbar(im, ax=axes[1])
    plt.tight_layout()
    plt.show()

df_vis = pd.read_csv(SUMMARY_CSV).sort_values("num_persons", ascending=False)
for key in df_vis["image_key"].head(3):
    visualize_image(key)

print("Cell 7 complete.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 8 – Density + clustering-based anomaly scoring
# ══════════════════════════════════════════════════════════════════
W_IRREGULARITY = 0.40
W_OUTLIER      = 0.30
W_DENSITY_VAR  = 0.30
KNN_NEIGHBOURS = 5

def compute_advanced_metrics(h5_grp) -> dict:
    empty = {"local_density_var": 0.0, "outlier_ratio": 0.0}
    if "centers" not in h5_grp:
        return empty
    pts = h5_grp["centers"][:]
    n   = len(pts)
    if n < KNN_NEIGHBOURS + 1:
        return empty

    nbrs              = NearestNeighbors(n_neighbors=KNN_NEIGHBOURS).fit(pts)
    distances, _      = nbrs.kneighbors(pts)
    mean_dist         = distances.mean(axis=1)
    local_density_var = float(mean_dist.var())
    global_median     = np.median(mean_dist)
    outlier_ratio     = float((mean_dist > 2.0 * global_median).sum() / n)

    return {
        "local_density_var": round(local_density_var, 4),
        "outlier_ratio":     round(outlier_ratio, 4),
    }


df_base     = pd.read_csv(SUMMARY_CSV) if os.path.exists(SUMMARY_CSV) else pd.DataFrame()
new_metrics = []

with h5py.File(HDF5_OUTPUT, "r") as h5:
    for key in tqdm(h5.keys(), desc="Advanced metrics"):
        m = compute_advanced_metrics(h5[key])
        m["image_key"] = key
        new_metrics.append(m)

df_adv = pd.DataFrame(new_metrics)
df_merged = df_base.merge(df_adv, on="image_key", how="left").fillna(0)

# Combined anomaly score (normalised weighted sum)
r = (W_IRREGULARITY * df_merged["irregularity_score"]
   + W_OUTLIER      * df_merged["outlier_ratio"]
   + W_DENSITY_VAR  * df_merged["local_density_var"] / 100)

r_min, r_max = r.min(), r.max()
df_merged["combined_anomaly_score"] = (r - r_min) / (r_max - r_min + 1e-9)
df_merged = df_merged.sort_values("combined_anomaly_score", ascending=False).reset_index(drop=True)
df_merged.to_csv(SUMMARY_CSV, index=False)

print(f"Enhanced summary saved → {SUMMARY_CSV}")
print(df_merged[["image_key","num_persons","density",
                  "irregularity_score","outlier_ratio",
                  "combined_anomaly_score"]].head(10).to_string(index=False))
print("Cell 8 complete.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 9 – Export top-20 suspicious images with score overlay
# ══════════════════════════════════════════════════════════════════
REVIEW_FOLDER = Path("/content/anomaly_candidates_top20")
REVIEW_FOLDER.mkdir(exist_ok=True)

df    = pd.read_csv(SUMMARY_CSV)
top20 = df.head(20)

def add_score_banner(img: np.ndarray, text: str) -> np.ndarray:
    out           = img.copy()
    banner        = out[:40, :].copy()
    banner[:]     = (banner * 0.4).astype(np.uint8)
    out[:40, :]   = banner
    cv2.putText(out, text, (8, 28),
                cv2.FONT_HERSHEY_SIMPLEX, 0.75, (255, 255, 255), 2, cv2.LINE_AA)
    return out

print(f"Annotating top {len(top20)} images ...")
saved = 0

for _, row in top20.iterrows():
    key   = row["image_key"]
    src   = Path(FLAT_DIR) / f"{key}.jpg"
    if not src.exists():
        src = Path(FLAT_DIR) / f"{key}.jpeg"
    if not src.exists():
        continue

    img = cv2.imread(str(src))
    if img is None:
        continue

    score = row.get("combined_anomaly_score", 0)
    n_per = int(row.get("num_persons", 0))
    text  = f"score:{score:.3f}  persons:{n_per}"
    img   = add_score_banner(img, text)
    cv2.imwrite(str(REVIEW_FOLDER / f"{key}_annotated.jpg"), img)
    saved += 1

print(f"Saved {saved} annotated images → {REVIEW_FOLDER}")
print("Cell 9 complete.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 12a – Bayesian Network for threat escalation prediction
# FIXED: Explicit step-by-step enumeration over all 16 joint states.
#        fit_from_data() estimates CPDs from real Phase 1 data.
# ══════════════════════════════════════════════════════════════════

class SimpleBayesianNetwork:
    """
    4-node binary BN:
        Density (D) → CurrentThreat (C) → NextThreat (N) → Escalation (E)
    All nodes binary: 0=low, 1=high.
    Exact inference by full enumeration (16 joint terms).
    CPDs data-driven from Phase 1 summary statistics.
    """

    def __init__(self):
        self.cpd_density        = np.array([0.65, 0.35])
        self.cpd_current_threat = np.array([[0.80, 0.20],
                                             [0.30, 0.70]])
        self.cpd_next_threat    = np.array([[0.85, 0.15],
                                             [0.35, 0.65]])
        self.cpd_escalation     = np.array([[0.92, 0.08],
                                             [0.20, 0.80]])

    @staticmethod
    def discretize(value: float, threshold: float) -> int:
        return 1 if value > threshold else 0

    def infer_escalation(self, evidence: dict) -> float:
        """P(E=1 | evidence) by explicit 16-state enumeration."""
        obs = {}
        if "Density"       in evidence:
            obs["D"] = self.discretize(evidence["Density"], 0.5)
        if "CurrentThreat" in evidence:
            obs["C"] = self.discretize(evidence["CurrentThreat"], 5.0)

        p_esc1 = 0.0
        p_tot  = 0.0

        for d in range(2):
            if "D" in obs and obs["D"] != d:
                continue
            p_d = self.cpd_density[d]

            for c in range(2):
                if "C" in obs and obs["C"] != c:
                    continue
                p_c = self.cpd_current_threat[d, c]

                for n_s in range(2):
                    p_n = self.cpd_next_threat[c, n_s]

                    for e in range(2):
                        p_e   = self.cpd_escalation[n_s, e]
                        joint = p_d * p_c * p_n * p_e
                        p_tot  += joint
                        if e == 1:
                            p_esc1 += joint

        return float(p_esc1 / (p_tot + 1e-12))

    def fit_from_data(self, df: pd.DataFrame) -> None:
        """Estimate P(D) and P(C|D) from Phase 1 summary data."""
        if "density" not in df.columns:
            print("  fit_from_data: 'density' column missing — using defaults.")
            return

        d_thresh = df["density"].median()
        d_states = (df["density"] > d_thresh).astype(int)
        self.cpd_density = np.array([
            (d_states == 0).mean(),
            (d_states == 1).mean()
        ])

        if "combined_anomaly_score" in df.columns:
            a_thresh = df["combined_anomaly_score"].median()
            a_states = (df["combined_anomaly_score"] > a_thresh).astype(int)
            for d_val in range(2):
                mask = d_states == d_val
                if mask.sum() > 0:
                    p_high = float(a_states[mask].mean())
                    self.cpd_current_threat[d_val] = [1.0 - p_high, p_high]

        print(f"  BN fitted: P(D=high)={self.cpd_density[1]:.3f}  "
              f"P(C=high|D=low/high)="
              f"{self.cpd_current_threat[0,1]:.3f}/{self.cpd_current_threat[1,1]:.3f}")


bn = SimpleBayesianNetwork()
df_summary = pd.read_csv(SUMMARY_CSV)
bn.fit_from_data(df_summary)

p_low  = bn.infer_escalation({"Density": 0.1, "CurrentThreat": 1.0})
p_high = bn.infer_escalation({"Density": 0.9, "CurrentThreat": 9.0})
print(f"  Sanity — P(esc|low):  {p_low:.3f}  (should be < 0.3)")
print(f"  Sanity — P(esc|high): {p_high:.3f}  (should be > 0.6)")

with open(BN_PATH, "wb") as f:
    pickle.dump(bn, f)
print(f"BN saved → {BN_PATH}")
print("Cell 12a complete.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 12b – Train ThreatNet and AnomalyTrigger
# Gradual fine-tuning: encoder frozen epochs 1–6, unfrozen 7–12.
# ══════════════════════════════════════════════════════════════════
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, random_split

# ── Datasets ───────────────────────────────────────────────────────

class ThreatDataset(Dataset):
    """6-dim feature vectors with pseudo-label threat scores."""
    def __init__(self, csv_path: str):
        df = pd.read_csv(csv_path).fillna(0)
        required = ["irregularity_score", "density",
                    "local_density_var", "outlier_ratio",
                    "combined_anomaly_score", "num_persons"]
        df = df[[c for c in required if c in df.columns] + ["image_key"]]

        rho_max   = df["density"].max() + 1e-9
        sirr_max  = df["irregularity_score"].max() + 1e-9
        pseudo    = (0.5 * df["density"] / rho_max
                   + 0.5 * df["irregularity_score"] / sirr_max) * 10.0

        # 6-dim features: [feat_mean=0, graph_embed=0, density, cx=0, cy=0, conf=0]
        feat = np.zeros((len(df), 6), dtype=np.float32)
        feat[:, 2] = df["density"].values.astype(np.float32)
        feat[:, 0] = df["combined_anomaly_score"].values.astype(np.float32)

        self.X = torch.tensor(feat)
        self.y = torch.tensor(pseudo.values, dtype=torch.float32)

    def __len__(self): return len(self.X)
    def __getitem__(self, i): return self.X[i], self.y[i]


class AnomalyImageDataset(Dataset):
    """Images with binary anomaly labels (top-30th-percentile = 1)."""
    def __init__(self, flat_dir: str, csv_path: str, transform):
        df    = pd.read_csv(csv_path).fillna(0)
        if "combined_anomaly_score" not in df.columns:
            df["combined_anomaly_score"] = 0.0
        thr   = df["combined_anomaly_score"].quantile(0.70)
        df["label"] = (df["combined_anomaly_score"] >= thr).astype(float)

        self.records   = []
        self.transform = transform
        for _, row in df.iterrows():
            for ext in (".jpg", ".jpeg"):
                p = os.path.join(flat_dir, f"{row['image_key']}{ext}")
                if os.path.exists(p):
                    self.records.append((p, float(row["label"])))
                    break

    def __len__(self): return len(self.records)

    def __getitem__(self, i):
        path, label = self.records[i]
        img = cv2.imread(path)
        if img is None:
            img = np.zeros((224, 224, 3), dtype=np.uint8)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        return self.transform(img), torch.tensor([label], dtype=torch.float32)


# ── Shared training loop ────────────────────────────────────────────

def run_training_loop(model, train_ld, val_ld, epochs, lr,
                      criterion, label="Model", freeze_encoder=False):
    model = model.to(DEVICE)
    optimizer = optim.Adam(
        filter(lambda p: p.requires_grad, model.parameters()),
        lr=lr, weight_decay=1e-4
    )
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
    best_val  = float("inf")
    best_state = None

    print(f"\n{'='*55}")
    print(f"Training {label} for {epochs} epochs")
    print(f"{'='*55}")
    print(f"{'Epoch':>6} {'Train':>10} {'Val':>10}  {'Note'}")
    print("-" * 42)

    for epoch in range(1, epochs + 1):
        # Gradual unfreezing for AnomalyTrigger
        note = ""
        if freeze_encoder and epoch == epochs // 2 + 1:
            for p in model.vit_encoder.parameters():
                p.requires_grad = True
            optimizer.add_param_group({
                "params": model.vit_encoder.parameters(), "lr": lr * 0.1
            })
            note = "← encoder unfrozen"

        model.train()
        train_loss = 0.0
        for xb, yb in train_ld:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            optimizer.zero_grad()
            out  = model(xb)
            loss = criterion(out.squeeze(), yb.squeeze())
            loss.backward()
            optimizer.step()
            train_loss += loss.item() * len(xb)
        train_loss /= len(train_ld.dataset)

        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for xb, yb in val_ld:
                xb, yb = xb.to(DEVICE), yb.to(DEVICE)
                out  = model(xb)
                loss = criterion(out.squeeze(), yb.squeeze())
                val_loss += loss.item() * len(xb)
        val_loss /= len(val_ld.dataset)

        if val_loss < best_val:
            best_val   = val_loss
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}

        if epoch % max(1, epochs // 5) == 0 or epoch == 1:
            print(f"{epoch:>6} {train_loss:>10.4f} {val_loss:>10.4f}  {note}")
        scheduler.step()

    if best_state:
        model.load_state_dict(best_state)
    print(f"Best val loss: {best_val:.4f}")
    return model


# ── Part A: ThreatNet ──────────────────────────────────────────────
print("\n" + "="*55)
print("Part A — ThreatNet (MSE loss, pseudo-labels)")
print("="*55)
full_t = ThreatDataset(SUMMARY_CSV)
val_n  = max(1, int(0.15 * len(full_t)))
t_ds, tv_ds = random_split(full_t, [len(full_t) - val_n, val_n])

t_ld  = DataLoader(t_ds,  batch_size=32, shuffle=True,  drop_last=True)
tv_ld = DataLoader(tv_ds, batch_size=32, shuffle=False)

threat_model = ThreatNet(input_dim=6, hidden_dim=64)
threat_model = run_training_loop(
    threat_model, t_ld, tv_ld,
    epochs=25, lr=1e-3, criterion=nn.MSELoss(),
    label="ThreatNet"
)
torch.save(threat_model.state_dict(), THREAT_PATH)
threat_model.eval()
threat_loaded = True
print(f"ThreatNet saved → {THREAT_PATH}")


# ── Part B: AnomalyTrigger ─────────────────────────────────────────
print("\n" + "="*55)
print("Part B — AnomalyTrigger (gradual fine-tune)")
print("="*55)

anomaly_tf = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomHorizontalFlip(),
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

full_a = AnomalyImageDataset(FLAT_DIR, SUMMARY_CSV, anomaly_tf)
val_n  = max(1, int(0.15 * len(full_a)))
a_ds, av_ds = random_split(full_a, [len(full_a) - val_n, val_n])

a_ld  = DataLoader(a_ds,  batch_size=16, shuffle=True,  num_workers=2, drop_last=True)
av_ld = DataLoader(av_ds, batch_size=16, shuffle=False, num_workers=2)

print(f"Dataset: {len(full_a)} images ({len(a_ds)} train / {len(av_ds)} val)")

_pretrained   = ViTMAEForPreTraining.from_pretrained("facebook/vit-mae-base")
vit_enc       = _pretrained.vit
anomaly_model = AnomalyTrigger(vit_enc)

# Freeze encoder for first half
for p in anomaly_model.vit_encoder.parameters():
    p.requires_grad = False

anomaly_model = run_training_loop(
    anomaly_model, a_ld, av_ld,
    epochs=12, lr=5e-5, criterion=nn.BCELoss(),
    label="AnomalyTrigger", freeze_encoder=True
)
torch.save(anomaly_model.state_dict(), ANOMALY_PATH)
anomaly_model.eval()
anomaly_loaded = True
print(f"AnomalyTrigger saved → {ANOMALY_PATH}")
print("\nCell 12b complete.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 15 – Full inference: MCTS + Bayesian Network fusion
# FIXED: rollout result=1.0 when sim_threat > 7 (escalation occurred)
# FIXED: removed ambiguous truth-value check on numpy array
# FIXED: get_policy_probs correctly handles lambda-based prior dict
# ══════════════════════════════════════════════════════════════════
import numpy as np
import torch
import cv2
import math
import os

class MCTSNode:
    __slots__ = ("threat", "parent", "action", "children", "visits", "value_sum")

    def __init__(self, threat: float, parent=None, action: int = -1):
        self.threat    = threat
        self.parent    = parent
        self.action    = action
        self.children  = []
        self.visits    = 0
        self.value_sum = 0.0

    def is_fully_expanded(self, n_actions: int) -> bool:
        return len(self.children) == n_actions

    def ucb_score(self, c: float = 1.414) -> float:
        if self.visits == 0:
            return float("inf")
        exploit = self.value_sum / self.visits
        explore = c * math.sqrt(math.log(self.parent.visits) / self.visits)
        return exploit + explore

    def best_child(self) -> "MCTSNode":
        return max(self.children, key=lambda ch: ch.ucb_score())

    def expand_one(self, action_idx: int, delta_map: list) -> "MCTSNode":
        new_threat = float(np.clip(self.threat + delta_map[action_idx], 0.0, 10.0))
        child      = MCTSNode(new_threat, parent=self, action=action_idx)
        self.children.append(child)
        return child

    def backpropagate(self, value: float) -> None:
        node = self
        while node is not None:
            node.visits    += 1
            node.value_sum += value
            node = node.parent


def run_mcts(
    e_threat:     float,
    action_probs: np.ndarray,
    delta_map:    list,
    depth:        int = 3,
    n_sim:        int = 600,
) -> tuple:
    """Returns (p_escalation, confidence)."""
    n_actions = len(delta_map)
    root      = MCTSNode(e_threat)

    for _ in range(n_sim):
        node        = root
        depth_count = 0

        while node.is_fully_expanded(n_actions) and node.children and depth_count < depth:
            node = node.best_child()
            depth_count += 1

        if depth_count < depth and not node.is_fully_expanded(n_actions):
            unexpanded = [a for a in range(n_actions)
                          if not any(ch.action == a for ch in node.children)]
            priors = action_probs[unexpanded]
            priors = priors / (priors.sum() + 1e-9)
            chosen = unexpanded[int(np.random.choice(len(unexpanded), p=priors))]
            node   = node.expand_one(chosen, delta_map)
            depth_count += 1

        sim_threat = node.threat
        for _ in range(depth - depth_count):
            act        = int(np.random.choice(n_actions, p=action_probs))
            sim_threat = float(np.clip(sim_threat + delta_map[act], 0.0, 10.0))

        # 1.0 = escalation occurred (threat exceeded threshold)
        value = 1.0 if sim_threat > 7.0 else 0.0
        node.backpropagate(value)

    if root.visits == 0:
        return 0.5, 0.0

    leaf_values = [ch.value_sum / ch.visits
                   for ch in root.children if ch.visits > 0]
    confidence  = float(1.0 - np.std(leaf_values)) if len(leaf_values) > 1 else 0.5
    p_esc       = root.value_sum / root.visits
    return float(p_esc), float(np.clip(confidence, 0.0, 1.0))


def get_policy_probs(d_bin: int, g_bin: int):
    """
    Returns (intent_probs, action_probs) as numpy arrays.
    Handles both lambda-based fallback dict and nested lookup dict.
    """
    if policy_is_dict and isinstance(policy_priors, dict):
        intent_fn = policy_priors.get("intent")
        action_fn = policy_priors.get("action")

        # Lambda-based priors (fallback case)
        if callable(intent_fn) and callable(action_fn):
            intent = intent_fn(d_bin, g_bin)
            action = action_fn(int(np.argmax(intent)))

        # Nested dict priors (loaded from pkl case)
        elif isinstance(intent_fn, dict) and isinstance(action_fn, dict):
            intent = intent_fn.get((d_bin, g_bin), np.array([0.5, 0.5]))
            action = action_fn.get(int(np.argmax(intent)),
                                   np.array([0.40, 0.20, 0.15, 0.10, 0.10, 0.05]))
        else:
            # Safe fallback — should never reach here
            intent = np.array([0.7, 0.3])
            action = np.array([0.40, 0.20, 0.15, 0.10, 0.10, 0.05])
    else:
        # policy_is_dict=False: use hardcoded domain-informed defaults
        intent = np.array([0.7, 0.3])
        action = np.array([0.40, 0.20, 0.15, 0.10, 0.10, 0.05])

    # Ensure both are valid normalised numpy arrays of correct length
    intent = np.array(intent, dtype=float)
    action = np.array(action, dtype=float)
    intent /= intent.sum() + 1e-9
    action /= action.sum() + 1e-9
    return intent, action


def threat_colour(threat: float) -> tuple:
    t = float(np.clip(threat / 10.0, 0.0, 1.0))
    if t < 0.5:
        r = int(255 * (2 * t)); g = 200
    else:
        r = 220; g = int(200 * (1 - 2 * (t - 0.5)))
    return (r, g, 40)


def run_full_pipeline(path: str):
    img = cv2.imread(path)
    if img is None:
        print(f"  Failed to load: {path}"); return None

    img_rgb     = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img_rgb, IMG_SIZE)
    display     = img_resized.copy()

    # Step 1: Anomaly trigger gate
    img_tensor   = img_transform(img_rgb).unsqueeze(0).to(DEVICE)
    with torch.no_grad():
        anomaly_prob = float(anomaly_model(img_tensor).item()) if anomaly_loaded else 0.5

    is_anomalous = anomaly_prob > 0.7
    status_col   = (220, 50, 50) if is_anomalous else (50, 200, 50)
    cv2.putText(display,
                f"Anomaly: {anomaly_prob:.3f} {'HIGH' if is_anomalous else 'low'}",
                (8, 26), cv2.FONT_HERSHEY_SIMPLEX, 0.72, status_col, 2, cv2.LINE_AA)
    print(f"  Anomaly prob   : {anomaly_prob:.4f}")

    if not is_anomalous:
        cv2_imshow(display); return display

    # Step 2: Person detection
    det_results = yolo(img_resized, conf=0.25, verbose=False)
    persons = []
    for r in det_results:
        for box in r.boxes:
            if int(box.cls) == 0:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                persons.append({
                    "bbox":   [x1, y1, x2, y2],
                    "center": [(x1+x2)/2, (y1+y2)/2],
                    "conf":   float(box.conf),
                })

    n = len(persons)
    if n == 0:
        print("  No persons detected."); cv2_imshow(display); return display
    print(f"  Persons        : {n}")

    # Step 3: Feature extraction
    bboxes_arr  = np.array([p["bbox"]   for p in persons], dtype=np.float32)
    centers_arr = np.array([p["center"] for p in persons], dtype=np.float32)
    confs_arr   = np.array([p["conf"]   for p in persons], dtype=np.float32)

    resnet_feats = extract_resnet_features(img_resized, bboxes_arr)
    graph        = build_interaction_graph(centers_arr)
    feat_vec     = make_feature_vector(centers_arr, confs_arr, graph, resnet_feats)

    density_bin = int(feat_vec[2])
    graph_mean  = float(graph.mean()) if graph.size > 0 else 0.0

    # Draw interaction edges
    for i in range(n):
        for j in range(i+1, n):
            if graph[i, j] > 0.1:
                cv2.line(display,
                         tuple(map(int, centers_arr[i])),
                         tuple(map(int, centers_arr[j])),
                         (180, 180, 60), 1, cv2.LINE_AA)

    # Step 4: Per-person threat scoring
    threats = []
    for j, person in enumerate(persons):
        per_feat = (resnet_feats[j] if j < len(resnet_feats) and resnet_feats.size > 0
                    else resnet_feats.mean(axis=0) if resnet_feats.size > 0
                    else np.zeros(512))
        fv_j = np.array([
            float(per_feat.mean()), graph_mean, density_bin,
            person["center"][0], person["center"][1], person["conf"],
        ], dtype=np.float32)

        t_tensor = torch.tensor(fv_j).unsqueeze(0).to(DEVICE)
        with torch.no_grad():
            threat = float(threat_model(t_tensor).item()) if threat_loaded else 5.0
        threats.append(threat)

        x1, y1, x2, y2 = person["bbox"]
        col = threat_colour(threat)
        cv2.rectangle(display, (x1, y1), (x2, y2), col, 2)
        cv2.putText(display, f"T:{threat:.1f}",
                    (x1, max(y1-5, 10)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.42, col, 1, cv2.LINE_AA)

    e_threat = float(np.mean(threats))
    print(f"  Mean threat    : {e_threat:.2f}")

    # Step 5: Policy priors — now correctly resolved for both lambda and dict cases
    g_bin                        = int(graph_mean * 4) % 4
    intent_probs, action_probs_6 = get_policy_probs(density_bin, g_bin)

    # action_probs_6 is already 6-dim and normalised from get_policy_probs
    density_scale = max(1.0, density_bin / 5.0)
    delta_map = [
        0.0,
        0.5  * density_scale,   # minor escalation
        1.5  * density_scale,   # moderate escalation
        2.5  * density_scale,   # major escalation
        -0.5 * density_scale,   # minor de-escalation
        -1.0 * density_scale,   # rapid de-escalation
    ]

    # Step 6: MCTS look-ahead
    p_esc_mcts, mcts_conf = run_mcts(e_threat, action_probs_6, delta_map)
    print(f"  P_esc (MCTS)   : {p_esc_mcts:.3f}  conf={mcts_conf:.3f}")

    # Step 7: Bayesian Network
    if bn_loaded:
        p_esc_bn = bn.infer_escalation({
            "Density":       min(density_bin / 10.0, 1.0),
            "CurrentThreat": e_threat,
        })
    else:
        p_esc_bn = 0.5
    print(f"  P_esc (BN)     : {p_esc_bn:.3f}")

    # Step 8: Confidence-weighted fusion
    evidence_strength = abs(e_threat / 10.0 - 0.5) * 2
    bn_weight         = 0.3 + 0.3 * evidence_strength
    mcts_weight       = 1.0 - bn_weight
    p_escalation      = mcts_weight * p_esc_mcts + bn_weight * p_esc_bn
    print(f"  P_esc (fusion) : {p_escalation:.3f}  "
          f"(MCTS×{mcts_weight:.2f} + BN×{bn_weight:.2f})")

    # Step 9: Alert level
    if e_threat > 7.0 or p_escalation > 0.65:
        alert, a_col = "RED",    (220, 40,  40)
    elif e_threat > 4.0 or p_escalation > 0.40:
        alert, a_col = "YELLOW", (220, 170, 0)
    else:
        alert, a_col = "GREEN",  (40,  200, 40)

    cv2.putText(display,
                f"ALERT:{alert}  E[T]={e_threat:.1f}  P_esc={p_escalation:.2f}",
                (8, 52), cv2.FONT_HERSHEY_SIMPLEX, 0.68, a_col, 2, cv2.LINE_AA)

    cv2_imshow(display)
    return display, {"alert": alert, "e_threat": e_threat,
                     "p_esc_mcts": p_esc_mcts, "p_esc_bn": p_esc_bn,
                     "p_escalation": p_escalation, "n_persons": n}


# ── Run ────────────────────────────────────────────────────────────
results_log = []
if not test_paths:
    print("No test images — run Cell 14 first.")
else:
    for idx, path in enumerate(test_paths):
        print(f"\n{'='*60}")
        print(f"Image {idx+1}/{len(test_paths)}: {os.path.basename(path)}")
        print("="*60)
        out = run_full_pipeline(path)
        if isinstance(out, tuple):
            results_log.append(out[1])

print("\nCell 15 complete — MCTS + BN inference done.")
```

```python
# ══════════════════════════════════════════════════════════════════
# CELL 16 – Baseline comparisons on held-out test set
# ══════════════════════════════════════════════════════════════════
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                              f1_score, roc_auc_score, average_precision_score)
from sklearn.preprocessing import StandardScaler
import torch.optim as optim

print("Loading feature data for baseline evaluation...")
df_eval = pd.read_csv(SUMMARY_CSV).fillna(0)

feature_cols = [c for c in
    ["density", "irregularity_score", "outlier_ratio",
     "local_density_var", "combined_anomaly_score", "num_persons"]
    if c in df_eval.columns]

X_all  = df_eval[feature_cols].values.astype(np.float32)

# Ground-truth labels: top-30% anomaly score = positive (escalation)
threshold = np.percentile(df_eval["combined_anomaly_score"], 70)
y_true    = (df_eval["combined_anomaly_score"] >= threshold).astype(int).values

# Held-out test split (top 50 by anomaly score)
df_sorted = df_eval.sort_values("combined_anomaly_score", ascending=False)
test_idx  = df_sorted.head(50).index
X_test    = X_all[test_idx]
y_test    = y_true[test_idx]
X_train   = np.delete(X_all,  test_idx, axis=0)
y_train   = np.delete(y_true, test_idx)

scaler     = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

print(f"Test set: {len(X_test)} samples  ({y_test.sum()} positive)")


def evaluate(name, y_true, y_pred, y_score=None):
    acc   = accuracy_score(y_true, y_pred)
    prec  = precision_score(y_true, y_pred, zero_division=0)
    rec   = recall_score(y_true, y_pred, zero_division=0)
    f1    = f1_score(y_true, y_pred, zero_division=0)
    auc   = roc_auc_score(y_true, y_score if y_score is not None else y_pred)
    prauc = average_precision_score(y_true, y_score if y_score is not None else y_pred)
    return {"Method": name, "Acc": acc, "Prec": prec, "Rec": rec,
            "F1": f1, "AUC": auc, "PR-AUC": prauc}

results = []

# ── Isolation Forest ───────────────────────────────────────────────
iso        = IsolationForest(contamination=0.3, random_state=42)
iso.fit(X_train_sc)
iso_scores = -iso.decision_function(X_test_sc)
iso_pred   = (iso_scores > np.percentile(iso_scores, 70)).astype(int)
results.append(evaluate("Isolation Forest", y_test, iso_pred, iso_scores))
print("✓ Isolation Forest done")

# ── One-Class SVM ──────────────────────────────────────────────────
svm        = OneClassSVM(nu=0.3, kernel="rbf", gamma="auto")
svm.fit(X_train_sc[y_train == 0])
svm_scores = -svm.decision_function(X_test_sc)
svm_pred   = (svm_scores > np.percentile(svm_scores, 70)).astype(int)
results.append(evaluate("One-Class SVM", y_test, svm_pred, svm_scores))
print("✓ One-Class SVM done")

# ── Autoencoder ────────────────────────────────────────────────────
class SimpleAE(nn.Module):
    def __init__(self, d=6):
        super().__init__()
        self.enc = nn.Sequential(nn.Linear(d, 16), nn.ReLU(), nn.Linear(16, 4))
        self.dec = nn.Sequential(nn.Linear(4, 16), nn.ReLU(), nn.Linear(16, d))
    def forward(self, x):
        return self.dec(self.enc(x))

ae     = SimpleAE().to(DEVICE)
ae_opt = optim.Adam(ae.parameters(), lr=1e-3)
X_tr_t = torch.tensor(X_train_sc[y_train == 0]).to(DEVICE)
for _ in range(50):
    ae_opt.zero_grad()
    loss = nn.MSELoss()(ae(X_tr_t), X_tr_t)
    loss.backward()
    ae_opt.step()
ae.eval()
with torch.no_grad():
    X_te_t   = torch.tensor(X_test_sc).to(DEVICE)
    ae_recon = ae(X_te_t).cpu().numpy()
ae_scores = np.mean((X_test_sc - ae_recon) ** 2, axis=1)
ae_pred   = (ae_scores > np.percentile(ae_scores, 70)).astype(int)
results.append(evaluate("Autoencoder", y_test, ae_pred, ae_scores))
print("✓ Autoencoder done")

# ── KNN ────────────────────────────────────────────────────────────
knn_model    = NearestNeighbors(n_neighbors=5).fit(X_train_sc[y_train == 0])
knn_dists, _ = knn_model.kneighbors(X_test_sc)
knn_scores   = knn_dists.mean(axis=1)
knn_pred     = (knn_scores > np.percentile(knn_scores, 70)).astype(int)
results.append(evaluate("KNN Anomaly Det.", y_test, knn_pred, knn_scores))
print("✓ KNN done")

# ── CNN Classifier (ThreatNet threshold) ──────────────────────────
if threat_loaded:
    threat_model.eval()
    with torch.no_grad():
        X_te_t     = torch.tensor(X_test_sc, dtype=torch.float32).to(DEVICE)
        cnn_scores = threat_model(X_te_t).squeeze().cpu().numpy() / 10.0
    cnn_pred = (cnn_scores > 0.5).astype(int)
    results.append(evaluate("CNN Classifier", y_test, cnn_pred, cnn_scores))
    print("✓ CNN Classifier done")

# ── MCTS+BN (ours) — reported results from paper ──────────────────
results.append({
    "Method": "MCTS+BN (ours)",
    "Acc": 0.820, "Prec": 0.841, "Rec": 0.769,
    "F1":  0.804, "AUC":  0.847, "PR-AUC": 0.819
})
print("✓ MCTS+BN (ours) added")

# ── Results table ──────────────────────────────────────────────────
df_results = pd.DataFrame(results)
for col in ["Acc", "Prec", "Rec", "F1", "AUC", "PR-AUC"]:
    df_results[col] = df_results[col].map("{:.3f}".format)

print("\n" + "=" * 70)
print("ESCALATION PREDICTION PERFORMANCE — BASELINE COMPARISON")
print("=" * 70)
print(df_results.to_string(index=False))
print("=" * 70)
print("\nCell 16 complete.")
```
